## 新聞內容找尋爬蟲來源

- 工具：Selenium、chromedriver(下載網址)
  chromedriver 建議找跟自己電腦chrome瀏覽器相同版本
- 爬取平台：Yahoo
- 演算法：
  1. query跟網頁標題利用模糊比對相似度
  2. query跟紅字(瀏覽器標出的關鍵字)計算比例
     - 註：若紅字比例很高，但標題相似度很低，代表可能只是轉發那篇文章
  3. 相似度不夠(fail_count)的次數3次後停止爬蟲(後續都不相干了)，若還停止，則會往下一頁繼續爬


```python
from selenium import webdriver
import time
import urllib
import os
from fuzzywuzzy import fuzz

URL_BASE = 'https://tw.search.yahoo.com/search?p='
query = '左手香的 8 種神奇功效：治腸胃、關節炎，還是乳腺癌剋星，勝過一堆保健食品！'

query_len = len(query)
print('query_len:', query_len)
print('-----------------------------------------------------')


# 啟動chrome瀏覽器
chromeDriver = r'C:\Users\User\webDriverTool\chromedriver'
driver = webdriver.Chrome(chromeDriver) 

# 最大化窗口
# driver.maximize_window()

# 瀏覽器打開爬取頁面
url = URL_BASE + query
driver.get(url)

url_list = [] # 所有找到的網址
view_count = 0 # 檢視過幾筆
success_count = 0 # 有效筆數
fail_count = 0 # 無效的筆數
page_count = 1 # 第幾頁
while(fail_count < 4):
    for element in driver.find_elements_by_class_name("algo"):
        title = element.find_element_by_class_name("ac-algo") # 網頁標題
        compTitle = element.find_element_by_class_name("compTitle") # 網址區域
        compText = element.find_element_by_class_name("compText") # 說明欄(包含紅色關鍵字)
        print('title:', title.text)

        # 模糊比對標題
        fuzz_ratio = fuzz.partial_ratio(query, title.text)
        print('fuzz_ratio:', fuzz_ratio)

        # 計算red_text_len
        red_text_len = 0
        red_text_list = compText.find_elements_by_css_selector('b')
        for text in red_text_list:
            print('text:', text.text)
            red_text_len = red_text_len + len(text.text)

        # 比對紅字跟query的比例
        print('text_ratio:', red_text_len/query_len)

        #如果標題文字相似度8成或說明文字相似度8成~12成 且 文字標題相似度要大於5成
        if ((fuzz_ratio>80) or (red_text_len/query_len>0.8 and red_text_len/query_len<1.2)) and (fuzz_ratio>50): 
            try:
                # 新的取網址方法
                ori_url = compTitle.find_element_by_class_name("ac-algo").get_attribute('href')
                print(ori_url)
                if ori_url not in url_list:
                    url_list.append(ori_url)
                success_count = success_count+1
            except:
                print('沒有直接網址')
        else:
            fail_count = fail_count+1
            print('相似度不夠')

        view_count = view_count+1
        print('-----------------------------------------------------')

        if fail_count>=3:
            print('已經超過可以錯誤的次數，強制結束')
            break
            
    if fail_count>= 3:
        break
    
    print('下一頁。到第', page_count+1, '頁')
    print('-----------------------------------------------------')
    next_page_element = driver.find_element_by_class_name("next")
    next_page_element.click()

driver.close()
print('共檢視幾筆資料(view_count):', view_count)
print('有效筆數(success_count)，可能包含重複的網址:', success_count)
print('所有網址(url_list):', url_list)
print('網址筆數:', len(url_list))
```
